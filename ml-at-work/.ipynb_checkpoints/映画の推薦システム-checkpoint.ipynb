MovieLensのデータ分析
--------------------

### ユーザーデータ読み込み

```python
import pandas as pd
import numpy as np
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('files/ml-100k/u.user', sep='|', names=u_cols)
users.head()
```

### 評価値情報の読み込み

```python
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('files/ml-100k/u.data', sep='\t', names=r_cols)
ratings['date'] = pd.to_datetime(ratings['unix_timestamp'],unit='s')
ratings.head()
```

```python
# 映画情報の読み込み
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
movies = pd.read_csv('files/ml-100k/u.item', sep='|', names=m_cols, usecols=range(5), encoding = "latin1")
movies.head()
```

```python
# 3つの情報をマージする
movie_rating = pd.merge(movies, ratings)
lens = pd.merge(movie_rating, users)
lens.head()
```

```python
# 最も評価された25作品のタイトル
lens.title.value_counts()[:25]
```

```python
# 評価の数と平均を集計し、平均が高い順に並べる
movie_stats = lens.groupby('title').agg({'rating': [np.size, np.mean]})
movie_stats.sort_values(by=[('rating', 'mean')], ascending=False).head()
```

- 評価件数が少ないため平均が高くなって上位にくるものが存在する

```python
# 100件以上の評価が高いもので評価値の平均が高い順に並べ替える
atleast_100 = movie_stats['rating']['size'] >= 100
movie_stats[atleast_100].sort_values(by=[('rating', 'mean')], ascending=False)[:15]
```

```python
# ヒストグラムを表示する
#from matplotlib import pyplot as plt
#%matplotlib inline

plt.style.use('ggplot')

# pandasの中にmatplotlibがあるので
lens.groupby('user_id').size().sort_values(ascending=False).hist()

plt.xlabel('rating size')
plt.ylabel('count of rating')
```

- テールに従って頻度が低くなるロングテールの分布となる
- これをジップの法則(Zipf's law)に従っているという

```python
# ユーザーごとの評価数と評価値の平均を出す
user_stats = lens.groupby('user_id').agg({'rating': [np.size, np.mean]})
user_stats['rating'].describe()
```

```python
from fastFM import als

from sklearn.feature_extraction import DictVectorizer
import numpy as np
train = [
    {"user": "1", "item": "5", "age": 19},
    {"user": "2", "item": "43", "age": 33},
    {"user": "3", "item": "20", "age": 55},
    {"user": "4", "item": "10", "age": 20},
]
v = DictVectorizer()
X = v.fit_transform(train)
print(X.toarray())
y = np.array([5.0, 1.0, 2.0, 4.0])
fm = als.FMRegression(n_iter=1000, init_stdev=0.1, rank=2, l2_reg_w=0.1, l2_reg_V=0.5)
fm.fit(X, y)
fm.predict(v.transform({"user": "5", "item": "10", "age": 24}))
```
