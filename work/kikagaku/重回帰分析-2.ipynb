入力変数と出力変数の切り分け
---------------------

```python
df.head(3)
```

```python
# df.iloc[行, 列]

# 全ての行を取る時
# df.iloc[:, 0]

# index指定
# x列まで取りたいとき
# x13のindexは12なので取りたい列のindex+1を指定する(指定したindexの前まで取得するということ)
# df.iloc[:, 0:13]
```

```python
df.iloc[0, 0]
```

```python
# xだけを切り出す
X = df.iloc[:,  :-1]
# yだけを切り出す
y = df.iloc[:,  -1:]
```

モデルの構築と検証
------------------

```python
from sklearn.linear_model import LinearRegression
```

```python
# モデルの宣言
model = LinearRegression()
```

```python
# モデルの学習
model.fit(X, y)
```

```python
# 検証(決定係数の計算)
model.score(X, y)
```

訓練データと検証データ
-------------------
例)受験勉強用10年分の過去問を購入

だめなケース

- 10年分で勉強(学習)
- 10年分の実力テスト(検証)

良いケース

- 前半の5年分で勉強(学習) ← 訓練データ
- 後半の5年分で実力テスト(検証) ← 検証データ

```python
from sklearn.model_selection import train_test_split
```

```python
# 訓練データと検証データの分割

# random_stateは乱数のシードを固定(再現性の確保)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
# モデルの学習
model.fit(X_train, y_train)
```

```python
# 検証 ← 検証データ
model.score(X_test, y_test)
```

```python
# 検証 ← 訓練データ
model.score(X_train, y_train)
```

- 訓練データの精度のほうが明らかに高いと過学習の可能性がある

予測値の計算
------------

```python
x = X.iloc[0, :]
```

```python
# 予測値の計算
y_pred = model.predict([x])[0]
y_pred
```
